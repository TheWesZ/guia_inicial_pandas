##Aprendendo a manipular o dataframe

1- Pegando uma coluna do dataframe

```python
produtos = vendas_df_import[['Produto', 'Valor Unitário']]
```

```python
produtos.head(10)
```

2- Pegando uma linha do dataframe

```python
vendas_df_import.loc[1]
```

2.1- Pegar linhas de acordo com uma condição

```python
lojas_selecionadas = vendas_df_import.loc[vendas_df_import['ID Loja'] == 'Iguatemi Esplanada']
lojas_selecionadas.head(10)
```

2.2- Pegar linhas e colunas especificas com o metodo loc[ ]

```python
#.loc[linhas, colunas]
lojas_selecionadas = vendas_df_import.loc[vendas_df_import['ID Loja'] == 'Iguatemi Esplanada', ['ID Loja', 'Produto', 'Quantidade']]
lojas_selecionadas.head(10)
```

2.3- Pegar um valor especifico com o metodo loc[ ]

```python
valor_especifico = vendas_df_import.loc[1, ['Produto']]
valor_especifico
```

#Calcular Indicadores

1. Value Counts

```python
transacoes_loja = vendas_df_import['ID Loja'].value_counts()
transacoes_loja.head(10)
```

2. Groupby

```python
faturamento_produto = vendas_df_import[['Produto', 'Valor Final']].groupby('Produto').sum()
faturamento_produto.head()
```
